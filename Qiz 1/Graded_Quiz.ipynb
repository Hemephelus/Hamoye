```python
import pandas as pd

url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'

food_df = pd.read_csv(url,encoding='latin-1')
food_df.describe(include='all')
```

```python
food_df
```

```python

area_2017 = food_df[['Area', 'Y2017']]
area_2017
```

```python
area_2017.groupby('Area').sum().sort_values(by=['Y2017'])
```

```python
Element_Code_and_Years_df = food_df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
print(Element_Code_and_Years_df.corr())
```

```python
madagascar = food_df[(food_df['Area'] == 'Madagascar')]
mad_protin = madagascar[['Element', 'Y2015']]
mad_protin.groupby('Element').sum()
```

```python
item_and_Years_df = food_df[['Item', 'Y2015', 'Y2018']]
item_and_Years_df.groupby('Item').sum()
```

```python
item_and_Years_df = food_df[['Element', 'Y2017']]
item_and_Years_df.groupby('Element').sum()
```
